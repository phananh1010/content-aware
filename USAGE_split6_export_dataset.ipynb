```python
import pickle
import os
import glob

import namespace
reload (namespace)

def create_filename_from_framepath(frame_filepath):
    tid, vid, fid = frame_filepath.split('/')[-4], \
                        frame_filepath.split('/')[-3], \
                            frame_filepath.split('/')[-1]
    name_template = '{}_{}_{}'#tid_vid_fid
    filename = name_template.format(tid, vid, fid)
    
    return filename

def copy_file(input_filepath, output_filepath):
    file_cmd_template = 'cp {} {}'
    file_cmd = file_cmd_template.format(input_filepath, output_filepath)
    return os.system(file_cmd)

def get_frames(vidpath, output_dirpath):
    #TODO: COPY the best versions of the frames of given videos
    #INPUT: path to video
    #OUTPUT: copy all frames to the designated directory
    
    framemask = os.path.splitext(vidpath)[0] + '/' + namespace.FRAMEVERSION_BEST + '/??????.jpg'
    for framepath in glob.glob(framemask): #go over all frame item in the best verion
        try:
            
            filename = create_filename_from_framepath(framepath)
            filepath = output_dirpath + '/' + filename
            print 'COPYING: {} \n\t TO: {}'.format(framepath, filepath)
            copy_file(framepath, filepath)
        except:
            print 'ERROR: ', framepath, vidpath
            raise
    return
```

```python
#print gt_dict.keys()
ID_list = [1, 7, 10, 15, 19, 23]
for ID in ID_list:
    GROUNDTRUTH_DICT_FILEPATH = namespace.GROUNDTRUTH_DICT_TEMPLATE.format(ID)
    gt_dict = pickle.load(open(GROUNDTRUTH_DICT_FILEPATH))
    for vid_path in gt_dict.keys():
        get_frames(vid_path, namespace.DIRPATH_EXPORT_DATASET)
```
